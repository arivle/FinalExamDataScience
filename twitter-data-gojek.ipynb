```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
from nltk.tokenize import word_tokenize 
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stop_factory = StopWordRemoverFactory().get_stop_words()
nltk.download('punkt')
```

```python
dataX = data.iloc[:, 0].values
dataY = data.iloc[:, 1].values
```

```python
def kataDasar (data):
    dataKataDasar=[]
    dataCopy = data
    data_size = dataCopy.shape[0]
    dictionary = ArrayDictionary(stop_factory)
    swr = StopWordRemover(dictionary)
    for i in range(data_size):
        token = nltk.tokenize.word_tokenize(swr.remove(stemmer.stem(re.sub(r"\d+", "", dataCopy[i])).translate(str.maketrans("","",string.punctuation))))
        dataKataDasar.append(token)
    return dataKataDasar
```

```python
dataxbaru = kataDasar(dataX)
dataxbaru
```

```python
dataybaru = dataY.tolist()
databaru = pd.DataFrame(list(zip(dataxbaru, dataybaru)), columns = ['kata', 'sentimen'])
databaru
```

```python
databaru['kata'] = [' '.join(map(str, l)) for l in databaru['kata']]
databaru
```

```python
count1 = Counter(" ".join(databaru[databaru['sentimen']==1]["kata"]).split()).most_common(100)
df1 = pd.DataFrame.from_dict(count1)
df1 = df1.rename(columns={0: "words in positive sentiment", 1 : "count"})
count2 = Counter(" ".join(databaru[databaru['sentimen']==0]["kata"]).split()).most_common(100)
df2 = pd.DataFrame.from_dict(count2)
df2 = df2.rename(columns={0: "words in negative sentiment", 1 : "count_"})
```

```python
from sklearn import feature_extraction
f = feature_extraction.text.CountVectorizer(stop_words = 'english')
x = f.fit_transform(databaru['kata'])
np.shape(x)
```

```python
y = databaru['sentimen'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=42)
print([np.shape(x_train), np.shape(x_test)])
```

```python
from sklearn import metrics
from sklearn import svm
list_C = np.arange(500, 2000, 100) #100000
score_train = np.zeros(len(list_C))
score_test = np.zeros(len(list_C))
recall_test = np.zeros(len(list_C))
precision_test= np.zeros(len(list_C))
count = 0
for C in list_C:
    svc = svm.SVC(C=C)
    svc.fit(x_train, y_train)
    score_train[count] = svc.score(x_train, y_train)
    score_test[count]= svc.score(x_test, y_test)
    recall_test[count] = metrics.recall_score(y_test, svc.predict(x_test))
    precision_test[count] = metrics.precision_score(y_test, svc.predict(x_test))
    count = count + 1 
```

```python
matrix = np.matrix(np.c_[list_C, score_train, score_test, recall_test, precision_test])
models = pd.DataFrame(data = matrix, columns = 
             ['C', 'Train Accuracy', 'Test Accuracy', 'Test Recall', 'Test Precision'])
models.head(n=10)
```

```python
m_confusion_test = metrics.confusion_matrix(y_test, svc.predict(x_test))
pd.DataFrame(data = m_confusion_test, columns = ['Predicted 0', 'Predicted 1'],
            index = ['Actual 0', 'Actual 1'])
```
